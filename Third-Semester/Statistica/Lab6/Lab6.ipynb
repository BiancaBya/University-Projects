```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend,xticks, plot 

data = norm.rvs(loc=165, scale=10, size=1000)
print(data)
```

```python

def functie(n, m, sigma):
    return norm.rvs(loc=m, scale=sigma, size=n)

data = functie(1000, 165, 10)

hist(data, bins=14, density = True,range=(130,200))
x = range(130,200)

plot(x,norm.pdf(x,loc=165,scale=10),'r-')
xticks(range(130,200,5))
grid()
show()
```

```python
sum([160 < x < 170 for x in data])/len(data)
print(norm.cdf(170,165,10) - norm.cdf(160,165,10))

print(data.mean())

```

```python
from scipy.stats import expon, uniform, bernoulli
from numpy import mean, std, multiply

i = bernoulli.rvs(p=0.4, size=1000)
data_ = [imp * expon.rvs(loc=0,scale=5) + (1-imp) * uniform.rvs(loc=4,scale=2) for imp in i]
print(sum(data_)/len(data_)) # valoarea medie

print(mean(data_))
print(std(data_)) # deviatia standard

dataFive = [i < 5 for i in data_]
print(mean(dataFive))


#probabilitatea teoretica
# sansa de a se intampla (0,4 sau 0,6) * probabilitatea pe acea ramura  
# P(T < 5) = 0,4 * P(T < 5 | I1) + 0,6 * P(T < 5 | I2)
#          expon.cdf(5, scale=5)        uniform.cdv(5, loc=4, scale=2)


print(0.4 * expon.cdf(5, scale=5) + 0.6 * uniform.cdf(5, loc=4, scale=2))
```

```python
from scipy.integrate import quad
from math import exp

g = lambda x : exp(-x**2)

x = [g(uniform.rvs(loc=-1, scale=4)) for _ in range(5000)]

print(mean(x) * 4)
print(quad(g, -1,3))
```
