```python
from scipy.stats import  bernoulli, binom
from matplotlib.pyplot import bar, show, hist, grid, legend, xticks


def deplasare(p: float, n: int):
    lista_deplasari = bernoulli.rvs(p, size=n)
    deplasari = [0]
    for depl in lista_deplasari:
        if depl == 1:
            deplasari.append(deplasari[len(deplasari) - 1] + 1)
        else:
            deplasari.append(deplasari[len(deplasari) - 1] - 1)
    return deplasari

print(deplasare(0.49, 100))
```

```python
from random import randrange
from scipy.stats import  bernoulli, binom
from matplotlib.pyplot import bar, hist, grid, show, legend



n = 10
p = 0.6



data = [deplasare(p, n)[-1] for _ in range(1000)]
bin_edges = [k+0.5 for k in range(-n-1, n+1)]
hist(data, bins=bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
# distribution = dict([(i,1/6) for i in range(1,7)])
# bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
# alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
xticks(range(-n, n+1))
show()
```

```python
from random import randrange
from scipy.stats import  bernoulli, binom
from matplotlib.pyplot import bar, hist, grid, show, legend



n = 10
p = 0.6



data = [deplasare(p, n)[-1] for _ in range(1000)]
bin_edges = [k+0.5 for k in range(-n-1, n+1)]
hist(data, bins=bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
distribution = dict([k, binom.pmf(k=(n+k)/2, n=n, p=p)] for k in range(-n, n+1))
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
xticks(range(-n, n+1))
show()
```

```python
n = 10
p = 0.6
N = 7 # nr noduri pe cerc


data = [deplasare(p, n)[-1] % N for _ in range(1000)]
bin_edges = [k+0.5 for k in range(-1, N)]
hist(data, bins=bin_edges, density = True, rwidth = 0.9, color = 'blue', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
# distribution = dict([k, binom.pmf(k=(n+k)/2, n=n, p=p)] for k in range(-n, n+1))
# bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
# alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
xticks(range(-1, N))
show()
```

```python
n = 10
p = 0.6
N = 7 # nr noduri pe cerc
prob_cerc = [0] * N
for k in range(-n, n+1):
    prob_cerc[k % N] += binom.pmf((n+k)/2, n=n, p=p)


data = [deplasare(p, n)[-1] % N for _ in range(1000)]
bin_edges = [k+0.5 for k in range(-1, N)]
hist(data, bins=bin_edges, density = True, rwidth = 0.9, color = 'blue', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
distribution = dict([k, prob_cerc[k]] for k in range(N))
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
xticks(range(N))
show()
```
