```python
from random import randint
from matplotlib.pyplot import plot, grid, title, show, yticks, xticks
from numpy import arange
from math import perm

# Aranj de 365 luate cate n/365 la puterea n = cazuri nefavorabile, deci 1 - asta
def simulari(n, nr_simulari):
    contor = 0
    for _ in range(nr_simulari):
        nr = []
        for i in range(n):
            numar = randint(1, 365)
            nr.append(numar)

        nr_egale = 0

        for i in range(n-1):
            for j in range(i + 1, n):
                if nr[i] == nr[j]:
                    nr_egale = 1
                    break

        contor += nr_egale

    prob = float(contor/nr_simulari)
    return prob



n = int(input("n = "))
simulari(n, 1000)

print(1 - perm(365,26)/pow(365,n))
```

```python
title('Plot test')
xs = range(2, 50)
ys = [simulari(x, 1000) for x in xs]
plot(xs,ys,'r*')
grid()
yticks(arange(0, 1, step=0.1))
xticks(range(0, 55, 5))
show()
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

nr_albastre = 0
axis('square')
axis((0, 1, 0, 1))
for _ in range(5000):
    F=[random(),random()]
    if dist(F,(0.5,0.5)) < 0.5:
        plot(F[0],F[1],'bo')
        nr_albastre += 1
    else:
        plot(F[0],F[1],'ro')
show()

print(nr_albastre/5000)
```
